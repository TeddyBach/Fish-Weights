```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn import preprocessing
import matplotlib
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
feature = pd.read_csv('https://raw.githubusercontent.com/ajn313/NYU2023SummerML1/main/Day5/fish_market_feature.csv')
label = pd.read_csv('https://raw.githubusercontent.com/ajn313/NYU2023SummerML1/main/Day5/fish_market_label.csv')
X = feature.values
y = label.values
length = feature["Length1"].values




#label.dropna() #dropping any empty/NA values so we don't get errors
#abel.head(100) #retrieving the first 100 values of each column
```

```python
X.shape, y.shape
```

```python
# Hint:
#split the dataset
from sklearn.model_selection import train_test_split
Xs = preprocessing.scale(X)
X_train, X_val, y_train, y_val = train_test_split(Xs, y, test_size=0.25, random_state=100)
#prediction
clf = Ridge()
clf.fit(X_train, y_train)


# Polynomial? Order? Would it be overfitting?
#Loss function
y_pred = clf.predict(X_val)
loss = mean_squared_error(y_val, y_pred)
print(loss)
# Plot?
plt.scatter(y_val, y_pred, color='b', label='Predicted')
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], color='r', linestyle='--', label='Ideal')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('Ridge Regression')
plt.legend()
plt.show()
# Go over the past slides and labs!
```
